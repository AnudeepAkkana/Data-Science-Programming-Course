```python
import random
```

```python
'''
This function gets creates a win_dict and the computer plays the value corresponding to the max key that has been played from the frequency_dict (human play)
'''

# Function to get computer choice
def RPS_computer(frequency_dict):
    computer_win_dict = {
        'R':'P',
        'P':'S',
        'S':'R'
    }
    maxval = max(list(frequency_dict.values()))
    max_list = []
    for k,v in frequency_dict.items():
        if v == maxval:
            max_list.append(k)
    computer_choice = random.sample([computer_win_dict[item] for item in max_list], k=1)[0]
    return computer_choice
```

```python
RPS_computer({'R':1, 'P':0, 'S':2})
```

```python
'''
This function allows the human to play R, P, S or Q. If the human enters anything else, it will return none
'''

# Function to get human choice
def RPS_human():
    human_choice = input('''Choices: 
    Press R for Rock, P for Paper, S for Scissors 
    Press Q to Quit''')
    if human_choice in ['R', 'P', 'S', 'Q']:
        return human_choice
    else:
        return None
```

```python
RPS_human()
```

```python
'''
This function determines who wins given the choices of both the computer and the human
'''

# Function to determine who wins
def determine_win(computer_play, human_play):
    human_win_dict = {
        'R':'P',
        'P':'S',
        'S':'R'
    }
    if computer_play == human_win_dict[f'{human_play}']:
        return 'Computer'
    elif human_play == human_win_dict[computer_play[0]]:
        return 'Human'
    else:
        return 'Draw'
```

```python
determine_win('R', 'P')
```

```python
if __name__=='__main__':

    flag = 0
    game_count = 0
    win_keeper = {
        'Human': 0,
        'Computer': 0,
        'Draw': 0
    }
    RPS_histogram = {
        'R': 0,
        'P': 0,
        'S': 0,
    }

    while flag == 0:

        # Round 1
        computer_play = RPS_computer(RPS_histogram)

        # Round 2
        human_play = RPS_human()

        if human_play == 'Q':
            flag = 1
            print(f'''{game_count} games were played \nHuman won {win_keeper['Human']} time(s)''')
        elif human_play == None:
            print(f'Game {game_count+1}: Error, please make a valid selection')
        else:
            RPS_histogram[human_play[0]] += 1
            result = determine_win(computer_play, human_play)
            if result == 'Draw':
                print(f'Game {game_count+1} result: {result}') 
            else:
                print(f'Game {game_count+1} result: {result} Wins!')
            win_keeper[result] += 1
            
        game_count += 1
```

```python
'''
This function makes the dictionary of keys being the moving window of size 3 of words in a tweet with values being 1. It uses the split function  
and list addition to get the result string
''' 
# Question 3a
def moving_window(twt):
    result_dict = {}
    for i in range(len(twt.split())):
        try:
            result = twt.split()[i]+' ' +twt.split()[i+1]+' '+twt.split()[i+2]
            result_dict[result]=1
        except:
            continue
    return result_dict
```

```python
moving_window('This tweet is awesome because I like data science')
```

```python
# Question 3b
import math

'''
Function to find the cosine similarity of two different tweets. Uses a loop over the words of the two tweets and adds 1 to matches everytime a window is matched
'''
def cosine(tweet1,tweet2):
    n1 = len(tweet1.keys())
    n2 = len(tweet2.keys())
    matches = 0
    for i in tweet1.keys():
        for j in tweet2.keys():
            if i==j:
                matches+=1
    cosine = matches / math.sqrt(n1*n2)
    return cosine
```

```python
cosine(moving_window("This tweet is awesome because I like data science"), moving_window("This tweet is not awesome because I don't like data science"))
```

```python
# Question 3c
with open('Santa.txt') as f:
    tweets = f.readlines()

dict_list = []
for tweet in tweets:
    tweet_dict = moving_window(tweet)
    dict_list.append(tweet_dict)

for i in dict_list:
    for j in dict_list:
        if i!=j:
            similarity = cosine(i,j)
            if similarity>0.5:
                print(f"tweet {dict_list.index(i)+1} and {dict_list.index(j)+1} are a near duplicate")
            else:
                print(f"tweet {dict_list.index(i)+1} and {dict_list.index(j)+1} are not a near duplicate")
```

```python
import re
import math
```

```python
# Reading the file, getting lines
with open('quotes.txt') as f:
    lines = f.readlines()

# Creating the quotes list
quotes = []
for i in range(len(lines)):
    if i%2 == 0:
        quotes.append(' - '.join((lines[i].strip(), lines[i+1].strip())))
```

```python
'''
This function uses regex to lower and only includes characters A-Z, a-z, 0-9 and underscores
'''

# Words from quotes
def split_q(quote):
    split_quote = re.split("[^A-Za-z0-9_]+", quote.lower())
    split_quote = [word for word in split_quote if word]
    return split_quote
```

```python
split_q(quotes[0])
```

```python
# Building the postings-list
postings_list = {}
for quote in quotes:
    words = split_q(quote)
    postings_value = {word:words.count(word) for word in words}
    postings_list[quote] = postings_value
```

```python
# Building the reverse posting list
rev_postings_list = {}
all_words = []
for quote in quotes:
    all_words += split_q(quote)
for word in all_words:
    rev_postings_value = {}
    for quote in quotes:
        if word in split_q(quote):
            rev_postings_value[quote] = split_q(quote).count(word)
    rev_postings_list[word] = rev_postings_value
```

```python
'''
This function takes a word and a quote as input and computes the tf and idf values. It uses the postings list and reverse postings list for each word respectivelly.
It proceeds to multiply the two to get the tfidf of that word and that quote. If the word inputted is not in the words of the quote, the tfidf is returned as 0.
'''

# Building a TF-IDF function
def TFIDF(word, quote):
    if word in split_q(quote):
        TF = postings_list[quote][word]/max(postings_list[quote].values())
        IDF = math.log(len(quotes)/len(rev_postings_list[word].keys()))
        TF_IDF = TF*IDF
    else:
        TF_IDF = 0
    return TF_IDF
```

```python
print(quotes[0])
TFIDF('how', quotes[0])
```

```python
'''
This function (input = word) returns the quote search dictionary with keys as the quotes and the values being the TFIDF of that word in the quotes
'''

# Building the quote search dictionary - input = 'word' in lowercase
def quote_search(word):
    qsearch_dict = {}
    for quote in rev_postings_list[word].keys():
        qsearch_dict[quote] = TFIDF(word, quote)
    return(qsearch_dict)
```

```python
quote_search('how')
```

```python
'''
This function is similar to quote_search but for multiple words. 
It takes a list of words to be searched as input. the values it returns are the sum of the tfidf values for the words in the quotes.
'''

# Building the multiple quote search dictionary - input = ['words'] all words should be in lowercase
def quote_search_mul(wordlist):
    mul_quote_search = {}
    quotes_with_words = []
    for word in wordlist:
        quotes_with_words += rev_postings_list[word].keys()

    sum_of_tfidfs = []
    for quote in quotes_with_words:
        sum_of_tfidfs.append(sum([TFIDF(word, quote) for word in wordlist if word in split_q(quote)]))
    for i in range(len(quotes_with_words)):
        mul_quote_search[quotes_with_words[i]] = sum_of_tfidfs[i]
    return mul_quote_search
```
